```python
import re
import random
import pickle
import argparse# подключаем библиотеки


class Generator_text:#создаем класс генерации текста
    
    def __init__(self,filename):
        text = open(filename, 'r', encoding='utf-8').read()#открываем файл с указанным путем и именем файла
        formatted_text = re.sub('[^а-яё,.!?]', ' ', text.lower())#форматируем текст (оставляем символы которые находят в квадратных скобках)
        self.splitted_text = re.split('\\s+', formatted_text)
        
    def fit(self):#создаем бигграммы по исходному тексту 
        result = {k: [] for k in self.splitted_text[:-1]}#создаем словарь для биграмм
        for i in range(len(self.splitted_text)-1):
            if self.splitted_text[i+1] not in result[self.splitted_text[i]]:
                result[self.splitted_text[i]].append(self.splitted_text[i+1])#добавляем в словарь биграммы
        
        with open('C:\\untitled2\\text_gen\\model.pkl', 'wb') as data:
            pickle.dump(result, data, protocol=pickle.HIGHEST_PROTOCOL)#сохраняем биграммы в файл
    
    def generate(self,length = 100):#генерация текста заданной длиной
        with open('C:\\untitled2\\text_gen\\model.pkl', 'rb') as data:
            bigrams = pickle.load(data)#извлечения биграмм из файла
        first_word = random.choice(list(bigrams.keys()))#берем случайное первое слово
        next_word = first_word
        generated = ''
        for i in range(length):#генерируем текст по биграммам
            generated += ''.join(next_word + " ")
            next_word = random.choice(bigrams[next_word])
            while next_word not in list(bigrams.keys()):
                next_word =  random.choice(list(bigrams.keys()))
        print(generated)
```

```python
def start():

    filename = "C:\\untitled2\\text_gen\\test.txt"
    length = 500
    text = Generator_text(filename)
    text.fit()
    text.generate(length)

start()
```
